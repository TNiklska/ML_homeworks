## Імпорт бібліотек

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

#Функції

```python
# Похідна функції
def pohidna_f (f):
    x = sp.symbols('x')
    derivative_f = sp.diff(f, x)

    print(f"Похідна  функції {f}:", derivative_f)
```

  1. З допомогою Python обчисліть похідну цієї функції.

###Розв'язок завдання 1

```python
#обчислюємо похідну
x = sp.symbols('x')
f = 10*x - 5
pohidna_f (f)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Визначення функції та її похідної
x = np.linspace(0.01, 10, 100)
f = 10*x-5
df = 10 + x*0 # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x-5', color='blue')
plt.plot(df, label="f'(x) = 10", color='green')



# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("Години на навчання")
plt.ylabel("Оцінки")
plt.legend()
plt.grid(True)

plt.show()
```

**Спостереження:**
Швидкість зміни оцінки залежить від кількості годин, витрачених на навчання, і оцінка збільшується пропорційно до  кількості витрачених годин.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



###Розв'язок завдання 2

```python

x = sp.symbols('x')
#Похідна f(x)=sin(x)
y = sp.sin(x)
pohidna_f (y)
```

```python
# Визначення функції та її похідної
x_vals = np.linspace(0, 5, 100)
f_vals = np.sin(x_vals)
df_vals = np.cos(x_vals) # Похідна f'(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label=r'$f(x) = \sin(x)$', color='blue')
plt.plot(x_vals, df_vals, label=r"$f'(x) = \cos(x)$", color='green')


# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = sp.symbols('x')
#Похідна f(x)=ln(x)
y = sp.log(x)
pohidna_f (y)
```

```python
# Визначення функції та її похідної
x_vals = np.linspace(0.01, 5, 100)
f_vals = np.log(x_vals)
df_vals = 1/(x_vals) # Похідна f'(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label=r'$f(x) = \log(x)$', color='blue')
plt.plot(x_vals, df_vals, label=r"$f'(x) = 1/(x)$", color='green')


# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

###Розвязок завдання 3

```python
x = sp.symbols('x')
#Похідна функції
y = x**2 - 5*x + 1
pohidna_f (y)
```

```python
# Визначення функції та її похідної
x_vals = np.linspace(0.01, 100, 1000)
f_vals = (x_vals)**2 - 5*(x_vals) + 1
df_vals = 2*(x_vals) - 5 # Похідна f'(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label= 'f(x) = x**2 - 5*x + 1', color='blue')
plt.plot(x_vals, df_vals, label= "f'(x) = 2x - 5", color='green')


# Додавання легенди, міток та заголовка
plt.title("Загальні витрати компанії від обсягу виробництва")
plt.xlabel('Кількість вироблених одиниць')
plt.ylabel('Загальні витрати компанії')
plt.legend()
plt.grid(True)

plt.show()
```
